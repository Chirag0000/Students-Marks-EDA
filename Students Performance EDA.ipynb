```python
df["meanscore"]=((df["math score"]+df["reading score"]+df["writing score"])/3).round()
df
```

```python
def gg(meanscore):
    if (meanscore=='Fail'):
        return 'F'
    if (meanscore<=30):
        return 'E'
    if (meanscore<=50):
        return 'D'
    if (meanscore<=60):
        return 'C'
    if (meanscore<=75):
        return 'B'
    if (meanscore<=99):
        return 'A'
df['grades']=df.apply(lambda x: gg(x['meanscore']), axis=1)
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
sns.countplot(df['math score'], palette = 'deep')
plt.title('Math Score',fontsize = 10)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['reading score'],palette='Set3')
plt.title('Reading Score', fontsize=10)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(df['writing score'],palette='prism')
plt.title('Writing Score', fontsize=10)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,5)
sns.violinplot(df['math score'],palette='muted')
plt.title('Math Score', fontsize=10)
plt.show()

plt.rcParams['figure.figsize']=(10,5)
sns.violinplot(df['reading score'],palette='Set2')
plt.title('Reading Score', fontsize=10)
plt.show()

plt.rcParams['figure.figsize']=(10,5)
sns.violinplot(df['writing score'],palette='Set1')
plt.title('Writing Score', fontsize=10)
plt.show()
```
